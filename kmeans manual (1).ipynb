```python
data['estimated_energy_used'] = data['Distance_Traveled_nm']/data['Efficiency_nm_per_kWh']
```

```python
features = ['estimated_energy_used',
            'Efficiency_nm_per_kWh']
# --- 1. PREPROCESSING: SCALING DATA ---
X = data[features].values

# Standarisasi (Z-score)
X_scaled = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
# --- 2. EUCLIDEAN DISTANCE FUNCTION ---

def euclidean_distance(X, centroid):
    return np.sqrt(((X - centroid) ** 2).sum(axis=1))


# --- 3. KMEANS MANUAL CLASS ---

class KMeansManual:
    def __init__(self, k=3, max_iter=100):
        self.k = k
        self.max_iter = max_iter
        self.centroids = None
        self.labels_ = None

    def fit(self, X):
        np.random.seed(42)
        self.centroids = X[np.random.choice(X.shape[0], self.k, replace=False)]

        for _ in range(self.max_iter):
            # Hitung jarak tiap titik ke semua centroid
            distances = np.array([euclidean_distance(X, c) for c in self.centroids])
            clusters = np.argmin(distances, axis=0)

            # Update centroid baru
            new_centroids = np.array([
                X[clusters == i].mean(axis=0) if np.any(clusters == i) else self.centroids[i]
                for i in range(self.k)
            ])

            # Cek konvergen
            if np.allclose(self.centroids, new_centroids, atol=1e-4):
                break
            self.centroids = new_centroids

        self.labels_ = clusters
        return self
```
